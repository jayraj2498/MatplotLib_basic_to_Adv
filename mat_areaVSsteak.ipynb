```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,5,6] 
y=[3,5,1,12,15] 
y2=[3,5,1,12,15] 
y3=[3,5,1,12,15] 


l=["area1","area2","area3"]
plt.stackplot(x,y,y2,y3 , labels=l )


plt.legend()
plt.show()
```

```python
x=[1,2,3,5,6] 
y=[3,5,1,12,15] 
y2=[3,5,1,12,15] 
y3=[3,5,1,12,15] 


l=["area1","area2","area3"]
plt.stackplot(x,y,y2,y3 , labels=l , colors=["r","g","y"])


plt.legend()
plt.show()
```

```python
x=[1,2,3,5,6] 
y=[3,5,1,12,15] 
y2=[3,5,1,12,15] 
y3=[3,5,1,12,15] 


l=["area1","area2","area3"]
plt.stackplot(x,y,y2,y3 , labels=l , colors=["r","g","y"] , baseline="sym")


plt.legend()
plt.title("Stack_plot")
plt.show()
```

```python

days = np.arange(1, 6)
component1 = [2, 4, 5, 7, 1]
component2 = [1, 3, 4, 6, 2]


plt.stackplot(days, component1, component2, labels=['Component 1', 'Component 2'])

plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Simple Stack Plot')
plt.legend()


plt.show()
```

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a stack plot with transparency
plt.stackplot(x, y1, y2, labels=['Sin', 'Cos'], alpha=0.7)

plt.xlabel('X-axis')
plt.ylabel('Values')
plt.title('Stack Plot with Transparency')
plt.legend()


plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = -np.cos(x)

plt.stackplot(x, y1, y2, labels=['Sin', '-Cos'])

plt.xlabel('X-axis')
plt.ylabel('Values')
plt.title('Stack Plot with Negative Values')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 2, 4, 3, 5]

# Create a step plot with horizontal and vertical steps
plt.step(x, y, where='mid', label='Midpoint Steps')
plt.step(x, [val + 1 for val in y], where='pre', label='Pre-point Steps')
plt.step(x, [val + 2 for val in y], where='post', label='Post-point Steps')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Horizontal and Vertical Steps')
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 2, 4, 3, 5]

# Create a filled step plot
plt.step(x, y, where='mid', label='Filled Steps', fillstyle='none', linestyle='--', color='orange')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filled Step Plot')
plt.legend()

# Display the plot
plt.show()
```
