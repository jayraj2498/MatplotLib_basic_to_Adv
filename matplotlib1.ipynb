```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 7]
c = ['r', 'y', 'g', 'b', 'm']

plt.bar(x, y, color=c)
plt.show()
```

```python
x= ['Category A', 'Category B', 'Category C', 'Category D'] 
rating = [25, 40, 30, 35] 

plt.bar(x,rating ,color="red")  

plt.xlabel("Categories")
plt.ylabel("values")
plt.title("Bar Plot Example")
plt.show() 

```

```python
x= ['Category A', 'Category B', 'Category C', 'Category D'] 
rating = [25, 40, 30, 35] 

plt.bar(x,rating ,color="y")  

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python
x= ['Category A', 'Category B', 'Category C', 'Category D'] 
rating = [25, 40, 30, 35] 

plt.bar(x,rating ,color="y" , width=0.1)  

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python



x= ['Category A', 'Category B', 'Category C', 'Category D'] 
rating = [25, 40, 30, 35] 

c = ['r', 'y', 'g', 'm'] 
plt.bar(x,rating ,color=c , width=0.2)  

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python
x= ['Cat A', 'Cat B', 'Cat C', 'Cat D'] 
rating = [25, 40, 30, 35] 

c = ['r', 'y', 'g', 'm'] 
plt.bar(x,rating ,color=c , width=0.5 , align='edge')  

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python
x= ['Cat A', 'Cat B', 'Cat C', 'Cat D'] 
rating = [25, 40, 30, 35] 

c = ['r', 'y', 'g', 'm'] 
plt.bar(x,rating ,color=c , width=0.5 , align='edge' , edgecolor='b' , linewidth=5 , linestyle=":",alpha=0.5)   

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python
x= ['Cat A', 'Cat B', 'Cat C', 'Cat D'] 
rating = [25, 27, 30, 20] 

c = ['r', 'y', 'g', 'm'] 
plt.bar(x,rating ,color=c , width=0.5 , label="Python")
plt.legend()   

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python
# double bar  plot 
```

```python
x= ['Cat A', 'Cat B', 'Cat C', 'Cat D'] 
y = [25, 27, 30, 20] 
z= [89, 51, 95, 55] 

# c = ['r', 'y', 'g', 'm'] 

plt.bar(x,y ,color='g' , width=0.2 , label="Python" , alpha=0.7)
plt.bar(x,z ,color='b' , width=0.5 , label="Java",alpha=0.7)
plt.legend()   

plt.xlabel("Categories" , fontsize=15)
plt.ylabel("values" , fontsize=15)
plt.title("Bar Plot Example")
plt.show()
```

```python

x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.bar(x, y, color='red', width=0.4, label="Python", align='center')
plt.bar(x, z, color='blue', width=0.4, label="Java", align='edge' )


plt.legend()

plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Bar Plot Example")

plt.show()
```

```python
x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.bar(x, y, color='red', width=0.4, label="Python", align='center', alpha=0.7)
plt.bar(x, z, color='blue', width=0.4, label="Java", align='edge', alpha=0.7 )


plt.legend()

plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Bar Plot Example")

plt.show()
```

```python

x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.bar(x, y, color='red', width=0.4, label="Python", align='center', alpha=0.7)
plt.bar(x, z, color='blue', width=0.4, label="Java", align='edge', alpha=0.7)


plt.legend()

plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Bar Plot Example")

plt.xticks( x, rotation=45)


plt.show()
```

```python
x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.barh(x, y, color='red',  label="Python", align='center', alpha=0.7)
plt.bar(x, z, color='blue',  label="Java", align='edge', alpha=0.7)


plt.legend()

plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Bar Plot Example")


plt.show()
```

```python



categories = ['A', 'B', 'C', 'D']
values = [25, 30, 20, 25]

plt.bar(categories, values, label='Category 1')
plt.bar(categories, [10, 15, 25, 20], bottom=values, label='Category 2')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Percentage Plot')
plt.legend()

plt.show()
```

```python
x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.barh(x, y, color='red',  label="Python", align='center', alpha=0.7)
plt.barh(x, z, color='blue',  label="Java", align='edge', alpha=0.7)


plt.legend()

plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Bar Plot Example")


plt.show()
```

```python
x = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
y = [25, 27, 30, 20]
z = [29, 21, 35, 25]

plt.bar(x, y, color='red', width=0.4, label="Python", align='center', alpha=0.7)
plt.bar(x, z, color='blue', width=0.4, label="Java", align='edge', alpha=0.7, bottom=y)


plt.legend()


plt.xlabel("Categories", fontsize=15)
plt.ylabel("Values", fontsize=15)
plt.title("Stacked Bar Plot Example")

plt.show()
```
