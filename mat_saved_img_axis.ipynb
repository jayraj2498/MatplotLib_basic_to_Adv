```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x= [10,30,40,50] 
y=['c' ,'java' ,'script','python'] 

c=['r','b','g','y']
plt.pie(x ,labels=y , colors=c , autopct="%0.1f%%") 



# plt.savefig("pie_chart_img")
plt.savefig("pie_chart_img" ,dpi=2000)   # to increse Quality of chart 
plt.show()
```

```python

x = [1, 2, 3, 4]
y1 = [2, 3, 1, 4]
y2 = [3, 1, 4, 2]


plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.scatter(x, y1, label='Scatter 1')
plt.legend()

plt.subplot(2, 2, 4)
plt.scatter(x, y2, label='Scatter 2')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()



plt.savefig("subplot.pdf")
plt.show()
```
