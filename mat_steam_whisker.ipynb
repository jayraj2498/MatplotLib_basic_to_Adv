```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y)
plt.show()
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y ,linefmt=':' )
plt.show()
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y ,linefmt=':' ,markerfmt="r*")
plt.show()
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y ,linefmt=':' ,markerfmt="r*" , basefmt="g")
plt.show()
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y ,linefmt=':' ,markerfmt="r*" , basefmt="g" , label="steamPlot")
plt.legend()
plt.show()
```

```python
x=[1,2,7,4,5] 
y=[3,4,5,6,7] 

plt.stem(x,y ,linefmt=':' ,markerfmt="r*" , basefmt="g" , label="steamPlot" , orientation="horizontal")
plt.legend()
plt.show()
```

# BOx vs Whisker 

```python
x=[10,20,70,40,50]   

plt.boxplot(x) 

plt.show()
```

```python
x=[10,20,70,40,50]   

plt.boxplot(x , notch=True) 

plt.show()
```

```python
x=[10,20,70,40,50]   

plt.boxplot(x , vert=False) 

plt.show()
```

```python
x=[10,20,70,40,50]   

plt.boxplot(x , widths=0.5) 

plt.show()
```

```python
x=[10,20,70,40,50]   

plt.boxplot(x , widths=0.5 , labels=["python"]) 

plt.show()
```

```python
x=[10,20,70,40,50]   

plt.boxplot(x , widths=0.5 , labels=["python"] , patch_artist=True) 

plt.show()
```
