```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]

# Create a simple pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'coral', 'lightpink'])

# Adding title
plt.title('Simple Pie Chart')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]

# Create an exploded pie chart
explode = (0, 0.1, 0, 0)  # Explode the second slice (Category B)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, colors=['skyblue', 'lightgreen', 'coral', 'lightpink'])

# Adding title
plt.title('Exploded Pie Chart')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels_outer = ['Category A', 'Category B']
sizes_outer = [60, 40]

labels_inner = ['Subcategory X', 'Subcategory Y', 'Subcategory Z']
sizes_inner = [30, 20, 10]

# Create a nested pie chart
fig, ax = plt.subplots()
ax.pie(sizes_outer, labels=labels_outer, autopct='%1.1f%%', startangle=90, radius=1,
       wedgeprops=dict(width=0.3, edgecolor='w'))

ax.pie(sizes_inner, labels=labels_inner, autopct='%1.1f%%', startangle=90, radius=0.7,
       wedgeprops=dict(width=0.3, edgecolor='w'), colors=['skyblue', 'lightgreen', 'coral'])

# Adding title
plt.title('Nested Pie Chart')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels_outer = ['Category A', 'Category B']
sizes_outer = [60, 40]

labels_inner = ['Subcategory X', 'Subcategory Y', 'Subcategory Z']
sizes_inner = [30, 20, 10]

# Create a nested pie chart
fig, ax = plt.subplots()
ax.pie(sizes_outer, labels=labels_outer, autopct='%1.1f%%', startangle=90, radius=1,
       wedgeprops=dict(width=0.3, edgecolor='w'))

ax.pie(sizes_inner, labels=labels_inner, autopct='%1.1f%%', startangle=90, radius=0.7,
       wedgeprops=dict(width=0.3, edgecolor='w'), colors=['skyblue', 'lightgreen', 'coral'])

# Adding title
plt.title('Nested Pie Chart')

# Display the plot
plt.show()
```
