### types of histograme

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , histtype='step')
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , histtype='stepfilled')
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , orientation='horizontal')
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , rwidth=0.7)  # for width 
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , rwidth=0.7 , log=True)  
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g' , edgecolor='red' , bottom=10 , rwidth=0.7 ,label="JavaSript")  
plt.legend()
plt.show()
```

```python
plt.xlabel('data')
plt.ylabel('range') 
plt.title("Histograme data") 




plt.hist(no ,color='g'  , bottom=10 ,label="JavaSript")  

plt.axvline(45, color='r' , label="AXVline")

plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 3  # Shifted distribution

# Create stacked histograms
plt.hist([data1, data2], bins=20, color=['skyblue', 'lightcoral'], edgecolor='black', stacked=True ,alpha=0.5)

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Stacked Histograms')

# Display the plot
plt.show()
```
