```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]

plt.scatter(x, y) 
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]


plt.scatter(x, y , color="r") 
plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]

c=['r','b','g','y','g']
plt.scatter(x, y , color=c) 
plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]

c=['r','b','g','y','g']
size=[10,45,500,20,50]
plt.scatter(x, y , color=c ,s=size) 
plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]

c=['r','b','g','y','g']
size=[10,45,500,20,50]
plt.scatter(x, y , color=c ,s=size ,marker="*") 
plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]

c=['r','b','g','y','g']
size=[10,45,500,20,50]
plt.scatter(x, y , color=c ,s=size , alpha=0.5) 
plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 7, 5, 9, 10]
z=  [22,33,44,55,66]

c=['r','b','g','y','g']
size=[10,45,500,20,50]

plt.scatter(x, y , color=c ,s=size ) 
plt.scatter(x, z , color=c ,s=size ) 

plt.title("scatter plot" , fontsize=15)
plt.xlabel("marks",fontsize=10)
plt.ylabel("marks" , fontsize=10)
plt.show()
```

```python
import numpy as np


x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 100, 50)  # Random sizes for markers


plt.scatter(x, y, s=sizes, c='green', alpha=0.8, label='Data Points')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Varying Size ')

plt.legend()
plt.show()
```
