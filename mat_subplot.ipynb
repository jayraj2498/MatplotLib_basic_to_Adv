```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

## fill_between


```python

x = [1, 2, 3, 4, 5]
y1 = [2, 1, 4, 3, 5]

plt.fill_between(x, y1,  color='skyblue', alpha=0.4, label='Filled Area')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Filled Area')
plt.legend()


plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y1 = [2, 1, 4, 3, 5]
y2 = [0, 0, 0, 0, 0]


plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Filled Area')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Filled Area')
plt.legend()


plt.show()
```

```python

x = [1, 2, 3, 4, 5]
y = [2, 1, 4, 3, 5]

plt.fill_between(x, y, y2=0, color='lightgreen', alpha=0.4, label='Filled Area with Baseline')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Filled Area with a Baseline')
plt.legend()


plt.show()
```

```python


x = [1, 2, 3, 4, 5]
y1 = [2, 1, 4, 3, 5]
y2 = [0, 0, 0, 0, 0]
y3 = [1, 2, 1, 4, 3]

plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Area 1')
plt.fill_between(x, y2, y3, color='lightcoral', alpha=0.4, label='Area 2')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Filled Areas')
plt.legend()


plt.show()
```

```python

x = [1, 2, 3, 4]
y1 = [2, 3, 1, 4]
y2 = [3, 1, 4, 2]


plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.scatter(x, y1, label='Scatter 1')
plt.legend()

plt.subplot(2, 2, 4)
plt.scatter(x, y2, label='Scatter 2')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()


plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y1 = [2, 3, 1, 4]
y2 = [3, 1, 4, 2]

# Create a custom 2x2 subplot grid
fig, axs = plt.subplots(2, 2, gridspec_kw={'width_ratios': [1, 2], 'height_ratios': [2, 1]})

# Plot line on the first subplot
axs[0, 0].plot(x, y1, label='Line 1')
axs[0, 0].legend()

# Plot scatter on the second subplot
axs[0, 1].scatter(x, y2, label='Scatter 2')
axs[0, 1].legend()

# Plot histogram on the third subplot
axs[1, 0].hist(y1, bins=3, label='Histogram 1')
axs[1, 0].legend()

# Plot bar chart on the fourth subplot
axs[1, 1].bar(x, y2, label='Bar 2')
axs[1, 1].legend()

# Adjust layout for better spacing
plt.tight_layout
```
